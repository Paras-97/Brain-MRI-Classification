```python
import math

def chunks(l, n):
    for i in range(0, len(l), n):
        yield l[i:i + n]

def mean(l):
    return sum(l) / len(l)

HM_SLICES = 20
compress_img=[]

chunk_sizes = math.ceil(len(new_img) / HM_SLICES)
for slice_chunk in chunks(new_img, chunk_sizes):
    slice_chunk = list(map(mean, zip(*slice_chunk)))
    compress_img.append(slice_chunk)

print(len(new_img), len(compress_img))
```

```python
for i in range(20):
    plt.imshow(compress_img[i])
    plt.show()
```
